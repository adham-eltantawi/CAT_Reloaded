## Q2.1

A fair die is rolled. What is the probability of rolling a number greater than 4?

```
Numbers greater than 4 on a fair die: {5, 6}  →  2 outcomes
Total outcomes: 6

P(X > 4) = 2/6 = 1/3
```

## Q2.2

You toss a fair coin three times:

- **(a)** What is the probability of three heads?
- **(b)** What is the probability that you observe exactly one heads?
- **(c)** Given that you have observed at least one heads, what is the probability that you observe at least two heads?

**Answer:**

```
Sample space (3 coin tosses): {HHH, HHT, HTH, HTT, THH, THT, TTH, TTT}  →  8 outcomes

(a) P(HHH) = 1/8

(b) Exactly one H: {HTT, THT, TTH}  →  3 outcomes
    P(exactly 1 head) = 3/8

(c) At least one H: all outcomes except TTT  →  7 outcomes
    At least two H: {HHH, HHT, HTH, THH}  →  4 outcomes
    P(at least 2 H | at least 1 H) = 4/7
```

## Q2.4 — Boxplot Analysis

Given a dataset of 800 measurements, the following statistics are known:

```
Range  = max - min = 41
Q1     = 114
Q3     = 126
IQR    = Q3 - Q1 = 12
```

> 126 is the Q3 on the boxplot (75th percentile) and 130 is the 80th percentile.
> The difference between them is 5%.
>
> → 0.05 × 800 = **40 measurements**

### (a) Lower and Upper Bounds

```
Lower Bound = Q1 - (1.5 × IQR) = 114 - (1.5 × 12) = 114 - 18 = 96
Upper Bound = Q3 + (1.5 × IQR) = 126 + (1.5 × 12) = 126 + 18 = 144
```

**Answer:**
- Lower Bound = **96**
- Upper Bound = **144**

## PS.1 — Plus One

> 🔗 [LeetCode: Plus One](https://leetcode.com/problems/plus-one/description/)

````python
class Solution(object):
    def plusOne(self, digits):
        i = len(digits) - 1

        while i >= 0 and digits[i] == 9:
            digits[i] = 0
            i -= 1

        if i >= 0:
            digits[i] += 1
        else:
            digits.insert(0, 1)

        return digits
````

## PS.2 — Missing Number

> 🔗 [LeetCode: Missing Number](https://leetcode.com/problems/missing-number/description/)

````python
class Solution(object):
    def missingNumber(self, nums):
        n = len(nums)
        expected = n * (n + 1) // 2
        actual = sum(nums)
        return expected - actual
````

## PS.3 — Bonus: Day of the Year

> 🔗 [LeetCode: Day of the Year](https://leetcode.com/problems/day-of-the-year/description/)

````python
class Solution(object):
    def dayOfYear(self, date):
        y, m, d = map(int, date.split('-'))

        days_in_month = [31, 28 + self.is_leap(y), 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

        return sum(days_in_month[:m-1]) + d

    def is_leap(self, year):
        return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)
````
